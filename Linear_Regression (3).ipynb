```python
#We have 0 in R&D Spend and Marketing spend columns so this may leads to overfitting
#We need to fill mean inplace of 0
for i in numerical_columns:
  if (df[i] == float(0)).any():
    df[[i]] = df[[i]].replace(0,df[[i]].median()).astype('float64')
```

```python
categorical_columns = df.select_dtypes(include='object').columns
```

```python
from sklearn.preprocessing import LabelEncoder # LabelEncoder is a class from the sklearn.preprocessing module in scikit-learn that converts categorical labels into a numeric form. Its commonly used in machine learning to handle categorical data, as many algorithms require numerical input.
le = LabelEncoder()
for i in categorical_columns:
  df[i] = le.fit_transform(df[i])
```

```python
df.head(3)
```

```python
X = df.drop('Profit',axis=1)
y = df['Profit']
```

```python
#Dataset is now ready for Spliting
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=38)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train =scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape,X_test.shape
```

```python
#Select the model and apply the model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print('MSE:',mean_squared_error(y_pred,y_test))
print('r2_score',r2_score(y_pred,y_test))
print('RMSE:',(np.sqrt(mean_squared_error(y_pred,y_test))))
```
